```python
data=pd.merge(new_movies1,credits,on='id')
```

```python
data
```

```python
import ast
#converting str to list
ast.literal_eval(data['genres'][0])
```

```python
data['genres']=data['genres'].apply(lambda x:ast.literal_eval(x))
data['cast']=data['cast'].map(lambda x:ast.literal_eval(x))
data['crew']=data['crew'].apply(lambda x:ast.literal_eval(x))
```

```python
data['title']=data['title'].str.lower()
```

```python
def make_list(x):
    gen=[]
    st=' '
    for i in x:
        if i.get('name')=='Science Fiction':
            gen.append('Sci-Fi')
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NaN
    else:
        return(st.join(gen))
```

```python
data1=data.copy()
```

```python
data1['genres']=data1.genres.map(lambda x:make_list(x))
```

```python
data1
```

```python
data1=data1.rename(columns={'title':'movie_title'})
```

#### getting actors name

```python
actors_name=[]
for i in range(len(data1)):
    cast=[]
    for j in data1['cast'][i]:
        cast.append(j.get('name'))
    actors_name.append(cast)
```

```python
actor_1_name=[]
for i in actors_name:
    if i!=[]:
        actor_1_name.append(i[0])
    else:
        actor_1_name.append('Unknown')
```

```python
actor_2_name=[]
for i in actors_name:
    if i!=[] and len(i)>2:
        actor_2_name.append(i[1])
    else:
        actor_2_name.append('Unknown')
```

```python
actor_3_name=[]
for i in actors_name:
    if i!=[] and len(i)>2:
        actor_3_name.append(i[2])
    else:
        actor_3_name.append('Unknown')
```

```python
data1['actor_1_name']=actor_1_name
data1['actor_2_name']=actor_2_name
data1['actor_3_name']=actor_3_name
```

```python
director=[]
for i in range(len(data1)):
    dirc=[]
    for j in data1['crew'][i]:
        if j.get('job')=='Director':
            dirc.append(j.get('name'))
    if dirc==[]:
        director.append('Unknown')
    else:
        director.append(dirc[0])
```

```python
data1['director_name']=director
```
