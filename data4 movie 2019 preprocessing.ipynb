```python
df=df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
df=df.drop(['Opening','Opening.1','Production company','Ref.'],axis=1)
```

```python
df.rename(columns={'Title':'movie_title'},inplace=True)
```

```python
df['movie_title']=df['movie_title'].str.lower()
```

```python
df
```

```python
#getting director
df['director_name']=df['Cast and crew'].str.split('(',expand=True)[0]
```

```python
df['actors']=df['Cast and crew'].str.split(';',expand=True)[2]
```

```python
#getting actor1
df['actor_1_name']=df['actors'].str.split(',',expand=True)[0]
#getting actor2
df['actor_2_name']=df['actors'].str.split(',',expand=True)[1]
#getting actor3
df['actor_3_name']=df['actors'].str.split(',',expand=True)[2]
```

```python
df.drop(['Cast and crew','actors'],axis=1,inplace=True)
```

```python
#getting genres
import tmdbv3api
from tmdbv3api import TMDb
import requests
import json
tmdb=TMDb()
tmdb.api_key='e770358dbe349592b1fba6cb607230fb'
```

```python
from tmdbv3api import Movie
tmdb_movie=Movie()
def get_genre(x):
    genres=[]
    result=tmdb_movie.search(x)
    movie_id=result[0].id
    response=requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json=response.json()
    if data_json['genres']:
        genre_str=' '
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            np.NaN
```

```python
df['genres']=df['movie_title'].map(lambda x:get_genre(str(x)))
```

```python
df['comb']=df['actor_1_name']+' '+df['actor_2_name']+' '+df['actor_3_name']+' '+df['director_name']+' '+df['genres']
```

```python
df=df[['movie_title','genres','director_name','actor_1_name','actor_2_name','actor_3_name','comb']]
```
